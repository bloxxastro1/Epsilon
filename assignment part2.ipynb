```python
import requests 
import csv
r = requests.get('http://universities.hipolabs.com/search')
data = r.json()
print (data)

with open('assignmet.csv', 'w', newline= '', encoding='utf-8') as file:

    writer = csv.DictWriter(file, fieldnames= ["country","name","web_pages"])
    writer.writeheader()

    for d in data:

       
        country = d["country"]
        name = d["name"]
        web_pages = d["web_pages"]
        writer.writerow({"country" : country ,"name" : name,"web_pages": web_pages})
```
